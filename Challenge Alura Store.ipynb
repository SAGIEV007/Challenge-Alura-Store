### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja1 = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja1.head()
```

#1. Análise do faturamento



```python
# Faturamento total de cada loja
faturamento1 = loja1['Preço'].sum()
faturamento2 = loja2['Preço'].sum()
faturamento3 = loja3['Preço'].sum()
faturamento4 = loja4['Preço'].sum()

print('Faturamento Loja 1:', faturamento1)
print('Faturamento Loja 2:', faturamento2)
print('Faturamento Loja 3:', faturamento3)
print('Faturamento Loja 4:', faturamento4)

# Média das avaliações
media_avaliacao1 = loja1['Avaliação da compra'].mean()
media_avaliacao2 = loja2['Avaliação da compra'].mean()
media_avaliacao3 = loja3['Avaliação da compra'].mean()
media_avaliacao4 = loja4['Avaliação da compra'].mean()

print('\nMédia de Avaliação Loja 1:', media_avaliacao1)
print('Média de Avaliação Loja 2:', media_avaliacao2)
print('Média de Avaliação Loja 3:', media_avaliacao3)
print('Média de Avaliação Loja 4:', media_avaliacao4)

# Custo médio do frete
frete1 = loja1['Frete'].mean()
frete2 = loja2['Frete'].mean()
frete3 = loja3['Frete'].mean()
frete4 = loja4['Frete'].mean()

print('\nCusto Médio do Frete Loja 1:', frete1)
print('Custo Médio do Frete Loja 2:', frete2)
print('Custo Médio do Frete Loja 3:', frete3)
print('Custo Médio do Frete Loja 4:', frete4)

# Produto mais e menos vendido
mais_vendido1 = loja1['Produto'].value_counts().idxmax()
menos_vendido1 = loja1['Produto'].value_counts().idxmin()

print('\nProduto mais vendido Loja 1:', mais_vendido1)
print('Produto menos vendido Loja 1:', menos_vendido1)
```

## Recomendação Final

Com base nas análises de faturamento, categorias mais populares, produtos mais e menos vendidos, média de avaliação e custo médio do frete, podemos agora formular uma recomendação sobre qual loja deveria ser vendida. Abaixo, será apresentada a justificativa para a escolha da loja a ser desinvestida, considerando os pontos fracos identificados em cada uma.

### Justificativa da Recomendação

Após analisar os principais indicadores de desempenho das quatro lojas, a **Loja 4** é a candidata recomendada para ser vendida. As razões para esta recomendação são as seguintes:

*   **Faturamento Total**: A Loja 4 apresentou o menor faturamento total entre todas as lojas, com R$ 1.384.497,58.

A Loja 1 se destaca com o maior faturamento R$ 1.534.509,12, e as Lojas 2 e 3 também superam a Loja 4. Um faturamento consistentemente mais baixo indica menor contribuição para a receita global da empresa.

*   **Média de Avaliação**: A Loja 4 teve uma média de avaliação de 3.995, que está na faixa intermediária, sendo ligeiramente superior à Loja 1 (3.976) mas inferior às Lojas 2 (4.037) e 3 (4.048). Embora não seja a pior em avaliações, não compensa o baixo faturamento com uma excelência marcante em satisfação do cliente.

*   **Custo Médio do Frete**: Curiosamente, a Loja 4 apresentou o menor custo médio de frete (R$ 31.27). Embora isso seja positivo em termos de eficiência logística, o impacto positivo do baixo custo de frete é ofuscado pelo seu faturamento total significativamente menor em comparação com as outras lojas. O volume de vendas e o preço dos produtos são fatores mais decisivos para a receita total.

Em resumo, apesar de ter um frete mais eficiente, a **Loja 4** demonstra o pior desempenho geral em termos de geração de receita, que é um dos indicadores mais críticos para a saúde financeira de uma unidade de negócio. Sua menor contribuição para o faturamento total, combinada com uma média de avaliação que não se destaca positivamente, a torna a candidata mais fraca para a retenção no portfólio da empresa. A venda da Loja 4 permitiria à empresa concentrar recursos nas lojas de maior desempenho e potencial.
